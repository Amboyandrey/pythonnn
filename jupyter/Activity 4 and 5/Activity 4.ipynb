```python
import matplotlib.pyplot as plt
import pandas as pd

bmi = pd.read_csv('bmi.csv')

first_10 = bmi.iloc[:10]
first_10
```

```python
female_data = first_10[first_10['Gender'] == 'Female']
male_data = first_10[first_10['Gender'] == 'Male']

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(range(5), female_data['Index'], color='blue', label='Female')
ax.bar(range(5, 10), male_data['Index'], color='red', label='Male')

ax.set_xlabel('Gender')
ax.set_ylabel('Index')
ax.set_title('Bar Plor Gender and Index for the First 10 Individuals')
ax.set_xticks(range(10))
ax.set_xticklabels(range(1, 11))
ax.legend()

plt.show()
```
